# Центральная предельная теорема своими руками

Распределение  хи-квадрат $\chi^2_k$,  имеющее натуральный параметр $k$, который называется числом степеней свободы. Пусть k=4.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
rv = sts.chi2(4)
sample = rv.rvs(1000) #Сгенерируйте из него выборку объёма 1000
```

```python
#постройте гистограмму выборки
plt.hist(sample, density=True, histtype='bar', ec='black', bins=30)
plt.xlabel('$x$')

# на том же графике построим теоретическую плотность распределения:
x = np.linspace(0,20,100)
pdf = rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf')
plt.legend()
plt.title("Гистограмма выборки и теоретическая плотность распределения")
```

Оценим распределение выборочного среднего вашей случайной величины при разных объёмах выборок [5,10,50]

Парметры нормального распределения, которыми, согласно центральной предельной теореме, приближается распределение выборочных средних, вычисляются по формулам. Параметры нормального распределения: $\mu = \mathbf{E}X = k, \sigma = \sqrt{\dfrac{\mathbf{D}X}{n}} = \sqrt{\frac{2k}{n}}$.

### Объем выборки 5

Параметры нормального распределения: $\mu = 4, \sigma = \sqrt{\frac{2*4}{5}}=1.2649110640673518 $

```python
average = [] # выборочных список средних 
for i in range(1000):
    sample = rv.rvs(5) #Сгенерируйте из него выборку объёма 5
    average.append(sum(sample)/5) # cчитаем среднее, как среднее арифмтическое выборки, и добавляем в список

#постройте гистограмму выборки
plt.hist(average, density=True, histtype='bar', ec='black')
plt.title("Гистограмма распределений выборочных средних для n=5 и, соответствующее им, нормальное распределение")

norm = sts.norm(4, np.sqrt(2*4 / 5)) # создаем случайную нормально распределенную величину с вычисленными парметрами
x = np.linspace(0, 20, 100)
plt.plot(x, norm.pdf(x))
plt.xlabel('x')
```

### Объем выборки 10

Параметры нормального распределения:  $\mu = 4, \sigma = \sqrt{\frac{2*4}{10}}=0.8944271909999159 $

```python
average = []
for i in range(1000):
    sample = rv.rvs(10) #Сгенерируйте из него выборку объёма 5
    average.append(sum(sample)/10)
#постройте гистограмму выборки
plt.hist(average, density=True, histtype='bar', ec='black')
plt.title("Гистограмма распределений выборочных средних для n=10 и, соответствующее им, нормальное распределение")

norm = sts.norm(4, np.sqrt(2*4 / 10)) # создаем случайную нормально распределенную величину
x = np.linspace(0, 20, 100)
plt.plot(x, norm.pdf(x))
plt.xlabel('x')
```
